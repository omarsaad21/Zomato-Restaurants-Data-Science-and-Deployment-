```python
str_count = df['rate'].value_counts().get('-', 0)
str_count
```

```python
df = df[df['rate'] != '-'].reset_index(drop=True)
```

```python
df['rate'].unique()
```

```python
nan_count = df['rate'].isna().sum()
nan_count
```

```python
# Extract the numeric part before '/5' and convert to float
df['rate_numeric'] = df['rate'].str.extract(r'(\d+\.?\d*)').astype(float)

# Calculate the mean of the numeric column
mean_rate = df['rate_numeric'].mean()

# Fill NaN values in the original 'rate' column with the calculated mean in 'rate_numeric' format
df['rate'] = df['rate'].fillna(f'{mean_rate}/5')
```

Check null values again

```python
nan_count = df['rate'].isna().sum()
nan_count
```

create a new column with the desired rate

```python
df['rate_flag'] = (df['rate_numeric'] > 3.75).astype(int)
```

```python
df = df.drop('rate_numeric', axis=1).reset_index(drop=True)
```

```python
df['approx_cost(for two people)'].unique()
```

```python
# Remove commas and convert to numeric
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '').astype(float)

# Fill missing values with the mean
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].mean())

```

```python
df.info()
```

```python
df = df.dropna(subset=['cuisines', 'rest_type']).reset_index(drop=True)
```

Distribution of Categorical features

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot for online_order
sns.countplot(data=df, x='online_order')
plt.title('Online Order Distribution')
plt.show()

# Bar plot for book_table
sns.countplot(data=df, x='book_table')
plt.title('Book Table Distribution')
plt.show()

# Bar plot for location (top 10 most frequent locations)
top_locations = df['location'].value_counts().head(10)
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.title('Top 10 Locations')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()

# Bar plot for cuisines (top 10 most frequent cuisines)
top_cuisines = df['cuisines'].value_counts().head(10)
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.title('Top 10 Cuisines')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot histograms for numeric columns
df[['votes', 'approx_cost(for two people)']].hist(bins=20, figsize=(10, 6))
plt.show()

```

Correlation analysis

```python
# Correlation matrix for numeric columns
corr_matrix = df[['votes', 'approx_cost(for two people)']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

 Location-wise Analysis
Check the average votes for restaurants in different locations:

```python
location_votes = df.groupby('location')['votes'].mean().sort_values(ascending=False).head(10)
location_votes.plot(kind='bar', figsize=(10, 6))
plt.title('Average Votes per Location')
plt.show()
```

Analyze if restaurants offering online orders have higher or lower costs:

```python
sns.boxplot(x='online_order', y='approx_cost(for two people)', data=df)
plt.title('Online Order vs. Approx Cost')
plt.show()
```

```python
sns.boxplot(x='book_table', y='votes', data=df)
plt.title('Book Table vs. Votes')
plt.show()
```

This chart shows the average cost for two people across different restaurant listing types (listed_in(type)):

```python
# Bar plot for avg cost by listed_in(type)
df.groupby('listed_in(type)')['approx_cost(for two people)'].mean().sort_values().plot(kind='barh', figsize=(10,6))
plt.title('Average Approx Cost by Listing Type')
plt.show()
```
